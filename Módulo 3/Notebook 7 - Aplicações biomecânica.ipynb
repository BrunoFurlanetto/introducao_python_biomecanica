# Taxa de mudança - Derivada

````python
# velocidade = mudança na posição / mudança no tempo
# mudança no tempo é o periodo de amostragem, que é o inverso da frequencia de amostragem
posicao = dados_cinematica['CdM_Y']
velocidade = np.diff(posicao) / ( 1 / frequencia_amostragem)

plt.figure(figsize=(10,3))
# alem da figura, podemos criar a area de plotagem (subplot), que nos da mais controle
ax = plt.axes()
ax.plot(velocidade)

# e novamente, os elementos graficos podem nos ajudar a analisar o sinal
ax.axhline(0, color='black', alpha=.2) # Eixo 0
````

### Comentário da célula anterior
* 1 / frequencia_de_amostragem nada mais é do que o tempo entre dois dados obtidos
* ```diff()``` faz a diferença entre os dados, par à par

````python
# para calcular a velocidade, perde-se um ponto do vetor, que foi usado para referencia
len(dados_cinematica),len(velocidade)
````

````python
# para facilitar o manuseio dos dados, é conveniente inserir um zero ao inicio do vetor de velocidades
velocidade = np.concatenate([ [0] , velocidade ])

len(velocidade)
````

### Comentário da célula anterior
* Aqui é importante tomar cuidado com como é feito isso, dessa forma aqui não é o ideal, porém é bem comum. O certo seria fazer a distância entre o ponto 1 e 3

````python
# podemos agora calcular a aceleração, a taxa com que a velocidade muda
aceleracao = np.diff(velocidade) / ( 1 / frequencia_amostragem)
len(aceleracao)
````

````python
# mais uma vez, nossa derivada tem um ponto a menos que o vetor original.
# Podemos inclui um 0 no final do vetor, para compensar o deslocamento temporal do calculo da velocidade
aceleracao = np.concatenate([ aceleracao, [0] ])
````

# Angulo bidimensional 

````python
# função para calcula de angulo bidimensional
def calcula_angulo_2D(a, b, c):
    # define os vetores (x,y) entre os pares de pontos
    ba = np.array(a) - np.array(b)
    bc = np.array(c) - np.array(b)
    # calcula o coseno e o seno do angulo formado entre os vetores
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    sine_angle = np.cross(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    # e em seguida define qual angulo corresponde ao coseno e seno obtido
    angle = np.arctan2(sine_angle, cosine_angle)
    # converte o angulo para o intervalo [0, 360)
    angle = np.degrees(angle)
    angle = angle if angle >= 0 else angle + 360
    # retorna o valor em graus
    return angle
````

### Comentário da célula anterior
* Pra cálculo de ângulo precisa de 3 valores

````python
# vamos testar a funcao
a = (1, 0) # Primeiro ponto e referência pra ele começar a contar no sentido anti horário
b = (0, 0) # Eixo de rotação
c = (0, 1) # Segundo ponto no qual tem o interesse pra poder encontrar o ângulo entre os dois pontos (a e b)

# construir visualização
fig, ax = plt.subplots()
xs = [a[0], b[0], c[0]]
ys = [a[1], b[1], c[1]]

ax.plot(xs, ys)

calcula_angulo_2D(a, b, c)
````

# suavização - convolução

````python
# vamos testar o comportamento de convolução entre diferentes sinais e filtros

sinal = np.concatenate([np.zeros(20), np.ones(20), np.zeros(20)]) + np.random.normal(0,.01,60)
# sinal = aceleracao

filtro = np.ones(4)/4

# realiza a convolução
sinal_filtrado = np.convolve(sinal, filtro)

# mostra a convolução
plt.figure(figsize=(8,4),dpi=200)
plt.plot(sinal,'.-',alpha=.2,label='sinal')
plt.plot(np.arange(len(filtro)),filtro,'.-',label='filtro')
plt.plot(sinal_filtrado, alpha=1, label='sinal filtrado')
plt.legend()
````

### Comentários da célula anterior
* Aqui da pra exemplificar o ruído da aceleração. Porque acaba aglotinando ruidos da posição e velocidade
* Explicar o que é convolução, soma da multiplicação de dois resultados entre duas funções e plota ele
* Sinal filtrado é a média movel nesse exemplo
* o ```np.random.normal()``` cria uma distribuição normal aleatóriamente ao redor de um valor

````python
# como uma convolução pode atrasar o sinal temporalmente, é importante atentar ao procedimento de inversao e refiltrgaem do sinal
# remove o excesso ao final do sinal filtrado
sinal_filtrado = sinal_filtrado[:len(sinal)]

# repete a convolução, invertendo o sinal e remove o excesso do final
sinal_filtrado_rev = np.convolve(sinal_filtrado[::-1], filtro)[:len(sinal)]
# inverte o resultado, para voltar ao original
sinal_filtrado_rev = sinal_filtrado_rev[::-1]

# mostra a convolução
plt.figure(dpi=200)
plt.plot(sinal, alpha=.2, label='Sinal original')
plt.plot(sinal_filtrado, alpha=.5, label='sinal atrasado')
plt.plot(sinal_filtrado_rev, alpha=.5, label='sinal sem atraso')
plt.legend()
````

### Comentários da célula anterior
* Pra não ter atraso a convolução é feita com o sinal invertido

````python
# podemos definir uma função com as etapas acima para reutiliza-la sem trabalho adicional no futuro
def media_movel_sem_atraso(sinal, janela):
    # define o filtro
    filtro = np.ones(janela)/janela
    # realiza a convolução
    sinal_filtrado = np.convolve(sinal, filtro)[:len(sinal)]
    # inverte, refiltra e reverte novamente
    sinal_filtrado_sem_atraso = np.convolve(sinal_filtrado[::-1], filtro)[:len(sinal)][::-1]

    return sinal_filtrado_sem_atraso
````

# Butterworth

````python
# Os filtros de resposta infinita ao impulso (IIR – Infinite Impulse Response) são semelhantes, 
# embora levem em conta as respostas anteriores do proprio filtro  

sinal = np.sin( np.linspace(0,2*np.pi*2,100) ) + np.sin( np.linspace(0,2*np.pi*20,100) )
# sinal = aceleracao
frequencia_amostragem = 100

# define os parametros e o filtro butterworth
b,a = signal.butter(N=4,Wn=3,btype='low',fs=frequencia_amostragem)

# aplica o filtro bidirecionalmente
sinal_filtrado = signal.filtfilt(b,a,sinal)

# mostra
fig,ax = plt.subplots(2,1,dpi=200)
ax[0].plot(sinal,label = 'sinal original')
ax[0].plot(sinal_filtrado, label = 'sinal filtrado')
ax[0].legend()
ax[1].plot(*signal.welch(sinal,fs=frequencia_amostragem),alpha=.5,label = 'sinal original')
ax[1].plot(*signal.welch(sinal_filtrado,fs=frequencia_amostragem),alpha=.5,label='sinal filtrado')
ax[1].legend()
````

# Detecção de evento

````python
# vamos trabalhar com dados de força de reação de solo para identificação de eventos em salto com contramovimento
dados_plataforma = pd.read_csv('https://raw.githubusercontent.com/conradotl/SBB_intro_prog/main/data_platforca.csv',sep=';',decimal=',') # Nesse arquivo o separador entre os dados é o ";"

# olha a estrutura dos dados
dados_plataforma.head()
````

````python
# vizualiza as forças
fig, ax = plt.subplots()
ax.plot(dados_plataforma['time'],dados_plataforma['fz'])
````

````python
# Como obter frequência de amostragem do equipamento
duracao = dados_plataforma['time'].values[-1] - dados_plataforma['time'].values[0]
amostras = len(dados_plataforma)
freq_amostragem = int(amostras/duracao)
freq_amostragem
````

````python
# uma vez que conhecemos minimamente o sinal disponivel, vamos seguir com a detecção de eventos.
# no caso, queremos descobrir os momentos em que o individuo toca e sai da plataforma
# podemos obter essa informação definindo um limiar de força que nos permita identificar se o sujeito esta com o peso no equipamento
limiar = 1

maior_limiar = dados_plataforma['fz'] > limiar

fig, ax = plt.subplots()
ax.plot(dados_plataforma['time'], dados_plataforma['fz'])
ax_t = plt.twinx(ax) # Fez um eixo novo gêmeo
ax_t.plot(dados_plataforma['time'], maior_limiar, color='black')
````

````python
# essa onda quadrada, aplicavel a diversas situações, pode ser utilizada para definir os momentos que o criterio 
# 1 - passa a ser verdadeiro
# 2 - passa a ser falso

auxiliar = np.diff(maior_limiar.astype(int))

fig,ax = plt.subplots()
ax.plot(dados_plataforma['fz'])
ax_t = plt.twinx(ax)
ax_t.plot(auxiliar,color='red')

ax.set_ylim(-2000,2000)
````

````python
# podemos identificar os momentos em que a derivada da onda quadrada é igual a 1 (saiu de False para True) e -1 (saiu de True para False)
inicios = np.where(auxiliar == 1)[0] 
finais = np.where(auxiliar == -1)[0]

# apresenta graficamente
plt.figure(figsize=(8, 3), dpi=300)
plt.plot(dados_plataforma['fz'])

for i in inicios:
    plt.axvline(i,ls=':', color='green')
    
for i in finais:
    plt.axvline(i,ls=':', color='red')
````
