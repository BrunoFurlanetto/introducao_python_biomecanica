## **Exercício 3**
Desenvolva um programa que leia o tempo (em segundos) de 3 voltas de um corredor e exiba o tempo total e a média por volta.

```Python
tempos = []

for i in range(3):
    tempo = float(input(f"Digite o tempo da volta {i+1} (em segundos): "))
    tempos.append(tempo)

tempo_total = sum(tempos)
media_volta = tempo_total / len(tempos)

print(f"Tempo total: {tempo_total} segundos")
print(f"Média por volta: {media_volta} segundos")
```

## **Exercício 4**
Crie um programa que receba a altura (em metros) e o peso (em kg) de um jogador de vôlei e calcule seu IMC (Índice de Massa Corporal).

```Python
altura = float(input("Digite a altura do jogador (em metros): "))
peso = float(input("Digite o peso do jogador (em kg): "))
imc = peso / (altura ** 2)
print(f"O IMC do jogador é: {imc:.2f}")
```

## **Exercício 7**
Crie um programa que leia o número de gols marcados por um jogador em 10 partidas e exiba a quantidade total de gols.
```Python
gols = [
    [3, 2, 1, 0, 2, 1, 4, 5, 2, 3],  # Gols do Jogador 1
    [1, 1, 2, 3, 4, 2, 1, 1, 2, 3],  # Gols do Jogador 2
    [2, 3, 3, 2, 1, 1, 2, 3, 4, 2],  # Gols do Jogador 3
    [0, 1, 1, 0, 2, 3, 2, 1, 0, 1]   # Gols do Jogador 4
]
```


```Python
gols = [
    [3, 2, 1, 0, 2, 1, 4, 5, 2, 3],  # Gols do Jogador 1
    [1, 1, 2, 3, 4, 2, 1, 1, 2, 3],  # Gols do Jogador 2
    [2, 3, 3, 2, 1, 1, 2, 3, 4, 2],  # Gols do Jogador 3
    [0, 1, 1, 0, 2, 3, 2, 1, 0, 1]   # Gols do Jogador 4
]

# Calculando o total de gols para cada jogador e o total geral da equipe
total_gols = [sum(jogador) for jogador in gols]
total_gols_equipe = sum(total_gols)

# Exibindo os resultados
print("Resumo dos Gols:")

for i in range(len(gols)):
    print(f"Jogador {i+1}: {total_gols[i]} gols")
    
print(f"Total de gols da equipe: {total_gols_equipe}")
```

## **Exercício 8**
Escreva um programa que receba a altura de 5 jogadores de basquete e determine a altura média do time e o nome do jogador mais alto.
```python
jogadores = {
    "Lucas": 1.95,
    "Carlos": 1.88,
    "Felipe": 2.03,
    "Matheus": 1.90,
    "Ricardo": 1.85
}
```


```Python
# Dicionário pré-definido com o nome e a altura dos jogadores
jogadores = {
    "Lucas": 1.95,
    "Carlos": 1.88,
    "Felipe": 2.03,
    "Matheus": 1.90,
    "Ricardo": 1.85
}

# Calculando a altura média do time
total_altura = sum(jogadores.values())
altura_media = total_altura / len(jogadores)

# Encontrando o jogador mais alto
jogador_mais_alto = max(jogadores, key=jogadores.get)
altura_jogador_mais_alto = jogadores[jogador_mais_alto]

# Exibindo os resultados
print("Resumo das Alturas:")
for nome, altura in jogadores.items():
    print(f"{nome}: {altura} m")

print(f"\nAltura média do time: {altura_media:.2f} m")
print(f"Jogador mais alto: {jogador_mais_alto} com {altura_jogador_mais_alto} m")
```

## **Exercício 10**
Crie um programa que receba a distância percorrida (em km) e o tempo gasto (em horas) por um ciclista e calcule sua velocidade média em km/h e m/s.

```Python
# Recebendo a distância percorrida e o tempo gasto
distancia_km = float(input("Digite a distância percorrida (em km): "))
tempo_horas = float(input("Digite o tempo gasto (em horas): "))

# Calculando a velocidade média em km/h
velocidade_kmh = distancia_km / tempo_horas

# Convertendo a velocidade para m/s
velocidade_ms = velocidade_kmh * 3.6

# Exibindo os resultados
print(f"\nVelocidade média:")
print(f"- Em km/h: {velocidade_kmh:.2f} km/h")
print(f"- Em m/s: {velocidade_ms:.2f} m/s")
```
