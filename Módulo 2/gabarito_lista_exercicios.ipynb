* **Exercício 1 - Cálculo de Velocidade Média**
Crie uma função chamada calcula_velocidade_media que receba a distância percorrida (em metros) e o tempo gasto (em segundos) por um atleta em uma corrida. A função deve retornar a velocidade média em metros por segundo (m/s).

```python
def calcula_velocidade_media(distancia, tempo):
    velocidade_media = distancia / tempo

    return velocidade_media

# Exemplo de uso:
velocidade = calcula_velocidade_media(100, 9.58)
print(f"Velocidade média: {velocidade:.2f} m/s")
```

* **Exercício 2 - Cálculo de Desempenho no Salto em Distância:**
Crie uma função que receba as distâncias de três saltos realizados por um atleta e retorne a média dos três saltos. Em seguida, crie outra função que classifique o desempenho do atleta com base na média obtida: "Excelente" para médias acima de 7 metros, "Bom" para médias entre 5 e 7 metros, e "Regular" para médias abaixo de 5 metros.

```python
def calcula_media_saltos(salto1, salto2, salto3):
    return (salto1 + salto2 + salto3) / 3

def classifica_desempenho(media_saltos):
    if media_saltos > 7:
        return "Excelente"
    elif 5 <= media_saltos <= 7:
        return "Bom"
    else:
        return "Regular"

# Exemplo de uso:
media = calcula_media_saltos(6.5, 7.2, 7.0)
classificacao = classifica_desempenho(media)
print(f"Média dos saltos: {media:.2f} m - Desempenho: {classificacao}")
```

* **Exercício 3 - Classificação de Atletas:**
Crie uma função que receba o número de gols marcados e assistências feitas por um jogador em uma temporada, e calcule um índice de desempenho usando a fórmula: desempenho = (gols * 4) + (assistências * 2). A função deve classificar o desempenho do jogador como "Alto", "Médio" ou "Baixo" com base no valor do índice.

```python
def calcula_desempenho(gols, assistencias):
    desempenho = (gols * 4) + (assistencias * 2)
    if desempenho > 20:
        return "Alto"
    elif 10 <= desempenho <= 20:
        return "Médio"
    else:
        return "Baixo"

# Exemplo de uso:
classificacao = calcula_desempenho(6, 3)
print(f"Classificação do desempenho: {classificacao}")
```

* **Exercício 4 - Calculadora de IMC para Atletas:**
Crie uma função que receba a massa corporal e a altura de um atleta e retorne o Índice de Massa Corporal (IMC). Em seguida, crie outra função que classifique o IMC em categorias como "Abaixo do peso", "Peso normal", "Sobrepeso" e "Obesidade", com base nos valores calculados.

```python
def calcula_imc(massa, altura):
    imc = massa / (altura ** 2)
    return imc

def classifica_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

# Exemplo de uso:
imc = calcula_imc(70, 1.75)
classificacao = classifica_imc(imc)
print(f"IMC: {imc:.2f} - Classificação: {classificacao}")
```

* **Exercício 5 - Análise de Rendimento em Provas de Corrida:**
Crie uma função que receba os tempos (em segundos) de um atleta em três corridas diferentes e retorne o tempo médio. Crie outra função que, com base no tempo médio, classifique o atleta como "Rápido" (tempo médio < 10s), "Moderado" (tempo entre 10s e 15s), ou "Lento" (tempo > 15s).

```python
def calcula_tempo_medio(tempo1, tempo2, tempo3):
    return (tempo1 + tempo2 + tempo3) / 3

def classifica_rendimento(tempo_medio):
    if tempo_medio < 10:
        return "Rápido"
    elif 10 <= tempo_medio <= 15:
        return "Moderado"
    else:
        return "Lento"

# Exemplo de uso:
tempo_medio = calcula_tempo_medio(9.8, 10.1, 9.5)
classificacao = classifica_rendimento(tempo_medio)
print(f"Tempo médio: {tempo_medio:.2f}s - Classificação: {classificacao}")
```

* **Exercício 6 - Cálculo de Pontuação em Tabelas de Futebol:**
Usando Pandas, crie um DataFrame com dados fictícios de 10 times de futebol, incluindo colunas como "Nome do Time", "Vitórias", "Empates", "Derrotas", "Gols Pró" e "Gols Contra". Em seguida, crie uma nova coluna chamada "Pontos", onde cada vitória vale 3 pontos, e cada empate vale 1 ponto. Calcule a pontuação total de cada time.

* **Exercício 9 - Análise Estatística de Provas Esportivas:**
Crie um DataFrame com os tempos de 10 atletas em três diferentes modalidades esportivas (corrida, natação, ciclismo). Calcule a média, mediana e desvio padrão dos tempos para cada modalidade. Em seguida, filtre os atletas que têm tempos abaixo da média em pelo menos duas modalidades.

```python
import pandas as pd

# Criando o DataFrame com tempos dos atletas
data = {
    'Atleta': ['Atleta1', 'Atleta2', 'Atleta3', 'Atleta4', 'Atleta5', 'Atleta6', 'Atleta7', 'Atleta8', 'Atleta9', 'Atleta10'],
    'Corrida': [12.5, 11.8, 13.0, 12.2, 11.9, 12.1, 13.3, 11.7, 12.6, 12.0],
    'Natação': [56.0, 55.5, 57.2, 55.8, 56.4, 56.1, 57.5, 55.9, 56.8, 56.2],
    'Ciclismo': [45.2, 44.8, 46.5, 45.0, 44.9, 45.1, 46.8, 44.7, 45.3, 45.0]
}

df = pd.DataFrame(data)

# Calculando a média, mediana e desvio padrão para cada modalidade
stats = pd.DataFrame({
    'Média': df[['Corrida', 'Natação', 'Ciclismo']].mean(),
    'Mediana': df[['Corrida', 'Natação', 'Ciclismo']].median(),
    'Desvio Padrão': df[['Corrida', 'Natação', 'Ciclismo']].std()
})

print("Estatísticas:")
print(stats)

# Calculando a média de cada modalidade
mean_values = df[['Corrida', 'Natação', 'Ciclismo']].mean()

# Filtrando os atletas que têm tempos abaixo da média em pelo menos duas modalidades
below_average = df[['Atleta']].copy()
below_average['Abaixo da Média'] = (
    ((df[['Corrida', 'Natação', 'Ciclismo']] < mean_values).sum(axis=1)) >= 2
)

filtered_atletas = below_average[below_average['Abaixo da Média']]

print("\nAtletas com tempos abaixo da média em pelo menos duas modalidades:")
print(filtered_atletas)
```

* **Exercício 10 - Simulação de Torneio de Futebol (desafio):**
Crie um DataFrame para simular um torneio de futebol com 10 times. O DataFrame deve ter as seguintes colunas:


* Nome do Time: Nome do time;
* Jogos Disputados: Número total de jogos (10 para cada time).
* Vitórias: Número de vitórias.
* Empates: Número de empates.
* Derrotas: Número de derrotas.
* Gols Pró: Gols marcados.
* Gols Contra: Gols sofridos.


* **Tarefas:**
    * Gere dados aleatórios para vitórias, empates, derrotas, gols pró e gols contra.
    * Certifique-se de que a soma de vitórias, empates e derrotas seja 10 para cada time.
    * Calcule o saldo de gols: Gols Pró - Gols Contra.
    * Calcule a pontuação de cada time: Vitórias * 3 + Empates * 1.
    * Classifique os times pela pontuação em ordem decrescente. Se houver empates na pontuação, use o saldo de gols para desempatar.
    * Exiba a tabela final com as seguintes colunas: Nome do Time, Jogos Disputados, Vitórias, Empates, Derrotas, Gols Pró, Gols Contra, Saldo de Gols, e Pontuação.

* **Dicas:**
    * Use pandas para criar e manipular o DataFrame.
    * Utilize random para gerar os dados aleatórios.

```python
import pandas as pd
import random

# Definindo o número de times
num_times = 10

# Função para gerar dados aleatórios para vitórias, empates e derrotas
def gerar_resultados():
    vitorias = random.randint(1, 7)
    empates = random.randint(0, 10 - vitorias)
    derrotas = 10 - (vitorias + empates)

    return vitorias, empates, derrotas

# Função para gerar dados aleatórios para gols
def gerar_gols():
    gols_pro = random.randint(10, 30)
    gols_contra = random.randint(10, 30)

    return gols_pro, gols_contra

# Gerando dados para os times
dados = {
    'Nome do Time': [f'Time{i+1}' for i in range(num_times)],
    'Jogos Disputados': [10] * num_times,
    'Vitórias': [],
    'Empates': [],
    'Derrotas': [],
    'Gols Pró': [],
    'Gols Contra': []
}

for _ in range(num_times):
    vitorias, empates, derrotas = gerar_resultados()
    gols_pro, gols_contra = gerar_gols()
    dados['Vitórias'].append(vitorias)
    dados['Empates'].append(empates)
    dados['Derrotas'].append(derrotas)
    dados['Gols Pró'].append(gols_pro)
    dados['Gols Contra'].append(gols_contra)

# Criando o DataFrame
df = pd.DataFrame(dados)

# Calculando o saldo de gols
df['Saldo de Gols'] = df['Gols Pró'] - df['Gols Contra']

# Calculando a pontuação (Vitórias = 3 pontos, Empates = 1 ponto, Derrotas = 0 pontos)
df['Pontuação'] = (df['Vitórias'] * 3) + (df['Empates'] * 1)

# Ordenando os times pela pontuação e pelo saldo de gols
df_sorted = df.sort_values(by=['Pontuação', 'Saldo de Gols'], ascending=[False, False])

# Exibindo a tabela final do torneio
print("Tabela Final do Torneio:")
print(df_sorted[['Nome do Time', 'Jogos Disputados', 'Vitórias', 'Empates', 'Derrotas', 'Gols Pró', 'Gols Contra', 'Saldo de Gols', 'Pontuação']])
```
